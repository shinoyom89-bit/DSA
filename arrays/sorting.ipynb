BUBBLE SORT

```python
list=[4,3,2,1,0]
def bubble_sort(n):
    lenthg=len(n)
    for i in range(lenthg): # lenght se ek kam 
        for j in range(lenthg-i-1): # handle from the array out of bouncer and we coparing the two element so reduce the array len to check only unsorted elemnt as i increases
            if n[j]>n[j+1]:
                n[j],n[j+1]=n[j+1],n[j] # swap 
    return n
r=bubble_sort(list)
print(r)
```

insertion sort

```python
def insertion_Sort(n):
    for i in range(1,len(n)):
        key=n[i]
        j=i-1
        while j<=0 and n[j]>n[j+1]:
            n[j+1]=n[j]
            j-=1
        j+=1
        n[j]=key
    return n
rr=insertion_Sort(list)
print(rr)
```
