BUBBLE SORT

```python
def bubble_sort(array:list)->list:
    length=len(array)
    for i in range(length): # input se ek kam
        for j in range(0,length-i-1):
            if array[j]>array[j+1]:
                array[j],array[j+1]=array[j+1],array[j]
    return array
listss=[4,523,234,23,1,3]
result=bubble_sort(listss)
print(result)
```

insertion sort

```python
def insertion_sort(arr:list)->list:
    length=len(arr)
    for i in range(1,length):  # input se ek kam due to zero based indexing
        key=arr[i]
        j=i-1
        while arr[j]>arr[j+1] and j>=0:
            arr[j],arr[j+1]=arr[j+1],arr[j]
            j-=1 # extra minus when the condition went wrong 
        arr[j+1]=key # to fix the extra minus we add to the right position 
    return arr
result=insertion_sort(listss)
print(result)
```
